```python
import pandas as pd
import numpy as np
import json
import re
from collections import defaultdict, OrderedDict

def extract_number(s):
    match = re.search(r'\d+$', s)  # ищем одну или более цифр в конце строки
    return str(match.group()) if match else None

def get_data_from_df(full_df, df):
    # Преобразование времени из часов в минуты и округление до целого
    # df['Время в минутах'] = np.ceil(df['Суммарное время операции в цехозаходе'] * 60).astype(int)

    # Формирование словаря для рабочих
    # Формирование словаря для рабочих с учетом типов операций для контролёров
    full_grouped_by_executor = full_df.groupby('Исполнитель')['Тип операции'].unique()
    full_grouped_by_controller = full_df.groupby('Контролёр')['Тип операции'].unique()

    full_workers_operations = {extract_number(executor): operations.tolist() for executor, operations in full_grouped_by_executor.items()}
    full_controllers_operations = {extract_number(controller): operations.tolist() for controller, operations in full_grouped_by_controller.items()}
    with open('controlers.txt', 'w') as file:
        for item in full_controllers_operations.keys():
            # Записываем каждый элемент списка в файл,
            # добавляя перевод строки после каждого элемента
            file.write("%s\n" % item)

    print("Список идентификаторов контролеров успешно сохранен в файл 'controlers.txt'")
    workers_data = {}
    grouped_workers = df.groupby('Исполнитель')['Тип операции'].unique()
    grouped_controllers = df.groupby('Контролёр')['Тип операции'].unique()
    # print(grouped_workers)
    for executor, operations in grouped_workers.items():
        worker_id = extract_number(executor)

        operation_types = full_workers_operations.get(worker_id, [])

        # Преобразуем типы операций в строки
        operation_types_str = [str(extract_number(op)) for op in operation_types]

        payment_type = df[df['Исполнитель'] == executor]['Вид оплаты'].iloc[0] if not df[df['Исполнитель'] == executor].empty else None
        workers_data[worker_id] = [operation_types_str, None, 1, payment_type, None]
    
    if len(grouped_controllers) > 0:
        for executor, operations in grouped_controllers.items():
            worker_id = extract_number(executor)

            operation_types = full_controllers_operations.get(worker_id, [])

            # Преобразуем типы операций в строки
            operation_types_str = [str(extract_number(op)) for op in operation_types]

            payment_type = df[df['Контролёр'] == executor]['Вид оплаты'].iloc[0] if not df[df['Контролёр'] == executor].empty else None
            workers_data[worker_id] = [operation_types_str, None, 1, payment_type, None]
    # print(workers_data)
    # Формирование словаря для операций
    jobs_data = {}
    operation_id_mapping = {}
    df_sorted = df.sort_values(by=['Обозначение ДСЕ', 'Номер операции в цехозаходе'])

    project_to_operations = defaultdict(list)
    operation_id_counter = 1  # Счетчик для порядковых номеров операций

    for index, row in df_sorted.iterrows():
        old_operation_id = extract_number(row['Описание операции'])
        operation_type = extract_number(row['Тип операции'])
        operation_time_minutes = int(np.ceil(row['Суммарное время операции в цехозаходе'] * 60))
        operation_time_minutes = int(max(operation_time_minutes, 1))  # Заменяем 0 на 1
        project_id = extract_number(row['Обозначение ДСЕ'])

        prev_op_id = None
        if project_to_operations[project_id]:
            prev_op_id = project_to_operations[project_id][-1]

        jobs_data[operation_id_counter] = [
            str(operation_type),
            operation_time_minutes,
            [prev_op_id] if prev_op_id is not None else [],
            1,
            operation_id_counter
        ]

        # Сохраняем соответствие старого и нового идентификаторов
        operation_id_mapping[old_operation_id] = operation_id_counter

        project_to_operations[project_id].append(operation_id_counter)
        operation_id_counter += 1

    # Формирование словаря для проектов
    project_data = {}
    for dse, group in df_sorted.groupby('Обозначение ДСЕ'):
        project_id = extract_number(dse)
        operation_ids = project_to_operations[project_id]
        project_data[project_id] = [operation_ids, 5000, group['Наряд-заказ'].iloc[0]]

    date_str = df['Дата задания'].dt.strftime('%Y-%m-%d').iloc[0]

    # Сериализация данных в JSON
    with open(f'data_from_df{date_str}.json', 'w', encoding='utf-8') as file:
        json.dump({
            'Machines': workers_data,
            'Operations': jobs_data,
            'Jobs': project_data
        }, file, ensure_ascii=False, indent=4, default=lambda o: int(o) if isinstance(o, np.integer) else o)
    with open(f'operation_id_mapping{date_str}.json', 'w', encoding='utf-8',) as file:
        json.dump({
            'OperationIdMapping': operation_id_mapping
        }, file, ensure_ascii=False, indent=4, default=lambda o: int(o) if isinstance(o, np.integer) else o)


    
```

```python
from collections import defaultdict, OrderedDict
import json
import pandas as pd
import numpy as np

def write_assignments_to_json(df):
    operations_data = defaultdict(list)
    structured_data = defaultdict(list)
    id_to_workers = OrderedDict()
    cumulative_time = defaultdict(int)
    cumulative_time_by_dse = defaultdict(lambda: defaultdict(int))
    cumulative_time_by_dse_only = defaultdict(int)
    date_str = df['Дата задания'].dt.strftime('%Y-%m-%d').iloc[0]
    # Загрузка mapping файла за пределами цикла для оптимизации
    with open(f'data_from_df{date_str}.json', 'r', encoding='utf-8') as file:
        machines = json.load(file)['Machines']
    
    with open(f'operation_id_mapping{date_str}.json', 'r', encoding='utf-8') as file:
        operation_id_mapping = json.load(file)["OperationIdMapping"]

    for index, row in df.iterrows():
        if pd.isna(row['Исполнитель']):
            executor = extract_number(row['Контролёр'])
        else:
            executor = extract_number(row['Исполнитель'])
        dse = int(extract_number(row['Обозначение ДСЕ']))
        operation_id = extract_number(row['Описание операции'])
        operation_time_minutes = row['Суммарное время операции в цехозаходе'] * 60

        # Сохраняем первичные данные операций для последующей сортировки
        operations_data[dse].append({
            'executor': executor,
            'operation_id': operation_id,
            'operation_time_minutes': operation_time_minutes,
            'op_id_mapped': operation_id_mapping.get(operation_id, None)
        })

    for dse, operations in operations_data.items():
        # Сортируем операции по идентификатору операции
        operations.sort(key=lambda x: x['op_id_mapped'])
        for op in operations:
            executor = op['executor']
            operation_time = int(np.ceil(op['operation_time_minutes']))
            if operation_time == 0:
                operation_time = int(max(operation_time, 1))
            start_time = max(cumulative_time[executor], cumulative_time_by_dse[dse][executor], cumulative_time_by_dse_only[dse])
            end_time = start_time + operation_time

            cumulative_time[executor] = end_time
            cumulative_time_by_dse[dse][executor] = end_time
            cumulative_time_by_dse_only[dse] = end_time
            structured_data[executor].append([dse, op['op_id_mapped'], start_time, end_time])

            for idx, machine_id in enumerate(machines.keys()):
                id_to_workers[machine_id] = idx

    solution_filename = f'solution_{date_str}.json'
    mapping_filename = f'id_to_workers_{date_str}.json'
    structured_data = OrderedDict(sorted(structured_data.items(), key=lambda item: list(id_to_workers.keys()).index(item[0])))
    print(structured_data)
    with open(solution_filename, 'w', encoding='utf-8') as file:
        json.dump(structured_data, file, ensure_ascii=False)
    
    with open(mapping_filename, 'w', encoding='utf-8') as file:
        json.dump(id_to_workers, file, ensure_ascii=False)
    
    print(f"Assignments written to {solution_filename} successfully.")
    print(f"Executor ID to list index mapping written to {mapping_filename} successfully.")
```

```python
def load_multiple_jsons(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    first_bracket_pos = content.find('[[')
    last_bracket_pos = content.rfind('[[[')
    first_list_str = content[first_bracket_pos:content.find(']]', first_bracket_pos) + 2]
    last_list_str = content[last_bracket_pos:content.rfind(']]]') + 3]
    first_list = json.loads(first_list_str)
    last_list = json.loads(last_list_str)
    result = {}
    for executor in first_list:
        result[executor[-1]] = last_list[executor[0]-1]
    result_file_path = file_path.replace('.txt', '_result.json')
    with open(result_file_path, 'w', encoding='utf-8') as f:
        json.dump(result, f, ensure_ascii=False, indent=4)

file_paths = ['scheduleRSC_89_1.txt', 'scheduleRSC_89_2.txt', 'scheduleRSC_234_1.txt', 'scheduleRSC_234_2.txt']
for file_path in file_paths:
    load_multiple_jsons(file_path)
```
